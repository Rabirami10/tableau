```python
fig = plt.figure(figsize=(20,6))
occupation_counts = data['Occupation'].value_counts().sort_values(ascending=False)
sb.countplot(x=data['Occupation'], order= occupation_counts.index, color='pink')
plt.show()
```

### Bivariate Analysis

```python
data.head()
```

```python
data.columns
```

```python
plt.bar(data['Quality of Sleep'], data['Stress Level'])
plt.xlabel('Sleep Quality')
plt.ylabel('Stress Level')
plt.show()
```

```python
plt.figure(figsize=(20,10))
sorted_data = data.sort_values('Sleep Duration', ascending=False )
plt.bar(sorted_data['Occupation'], sorted_data['Sleep Duration'], color='c')
plt.ylabel("Sleep Duration")
plt.xlabel("Occupation")
plt.title('Sleep Duration Based on Occupation')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sorted_data = data.sort_values('Sleep Duration', ascending=False )
plt.bar(sorted_data['Sleep Disorder'], sorted_data['Sleep Duration'], color='y')
plt.ylabel("Sleep Duration")
plt.xlabel("Sleep Disorder")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sb.boxenplot(x=data['BMI Category'], y=data['Sleep Duration'], palette="viridis")
plt.title('Sleep Duration By BMI Category')
plt.show()
```

```python
sb.barplot(x = data['Gender'], y = data['Sleep Duration'], palette='plasma')
plt.show()
```

```python
sb.catplot(x='Age',y='Heart Rate', data = data, kind="point", aspect=2, height=6)
plt.title('Heart Rate Variation By Age')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sb.swarmplot(x=data['Occupation'],y=data['Stress Level'], palette="pastel")
plt.title("Stress Level By Occupation")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sb.lineplot(x=data['Age'], y= data['Sleep Duration'], marker="*", markersize=8)
plt.title('Sleep Duration Variation by Age')
plt.show()
```

```python
plt.scatter(data['Physical Activity Level'], data['Quality of Sleep'])
plt.title('Physical Activity Level vs. Quality of Sleep')
plt.xlabel('Physical Activity Level')
plt.ylabel('Quality of Sleep')
plt.show()
```

```python
occupation = sb.FacetGrid(data, col="Occupation", col_wrap=3, height=4)
occupation.map(sb.lineplot, "Age", "Sleep Duration", marker='o',markersize=6)
occupation.set_titles(col_template="{col_name}")
occupation.set_axis_labels("Age", "Sleep Duration")
occupation.fig.suptitle("Sleep Duration Variation by Age and Occupation", y=1.02)
plt.show()
```

```python
sb.pairplot(data)
plt.show()
```

## Multivariate Analysis

```python
sb.violinplot(x = data['Sleep Disorder'], y = data['Quality of Sleep'], hue= data['Gender'], split=True)
plt.show()
```

```python
plt.figure(figsize=(15,6))
sb.distplot(data[data['Sleep Disorder']== 'None']['Age'], label="None")
sb.distplot(data[data['Sleep Disorder']== 'Sleep Apnea']['Age'], label="Sleep Apnea")
sb.distplot(data[data['Sleep Disorder']== 'Insomia']['Age'], label= "Insomia")
plt.title("Distplot for Sleep Disorder with Age")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,6))
sb.scatterplot(x=data['Stress Level'], y=data['Physical Activity Level'], size= data['Sleep Duration'])
plt.title("Scatter Plot of Sleep Duration, Stress Level, and Physical Activity")
plt.show()
```

```python
sb.heatmap(data.corr(),  annot= True)
plt.show()
```
